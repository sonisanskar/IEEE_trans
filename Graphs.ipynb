# COMPARISON OF MODEL DURING TRAINING

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
M=['dataset-1(b)','dataset-2(b)','dataset-5(m)','dataset-6(m)']*3

y=[0.656622698,0.460466564,0.101568267,0.07502009]
y1=[0.734262347,0.525866628,0.186243355,0.086125955]
y2=[0.664307773,0.484426469,0.172511935,0.102311589]
Y=y+y1+y2
name=["Text_6"]*4+['Text_4']*4+['Yoon-kim']*4

data={"Model":name,"val_loss":Y,'M':M}
df=pd.DataFrame(data)
my_plot = sns.lineplot(data=df,x="M", y="val_loss",hue="Model",sort=False);
plt.legend(loc='upper right')
plt.savefig('saving-a-seaborn-plot-as-pdf-file-300dpi.pdf', dpi=300)
my_plot.set(xlabel ='DATASETS', ylabel ='val_loss') 

plt.title('MODEL_VAL_LOSS'); 
plt.savefig('saving-a-seaborn-plot-as-pdf-file-300dpi2loss.pdf', dpi=300)
plt.show()
```
